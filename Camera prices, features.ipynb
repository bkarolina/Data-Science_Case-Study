I start in reverse order as in the title (which makes sense, because right now we have nothing just this nice text above). I highly recommend this video (https://www.youtube.com/watch?v=3xQTJi2tqgk), personally it helps me a lot to find a way around.

```python
# Load libraries
import requests
from bs4 import BeautifulSoup
```

```python
# Acces our desirable website
r = requests.get ("https://www.dpreview.com/products/cameras/all?page=1")
# Put it into the soup
soup = BeautifulSoup(r.content)
```

```python
# STEP 1 : have a list of camera's models
# Find models of cameras
for link in soup.find_all("div", {"class" : "name"}):
    print(link.text)
```

Weheeeej, now we have all cameras models from page 1 search. 

```python
# STEP 1 : have a list of camera's price

for link in soup.find_all("div", {"class" : "prices"}):
    print(link.text)
```

```python
#STEP 2 Get the links to the each models
for link in soup.find_all("a"):
    print(link.get("href"))
```

This is not want I wanted. I am getting some URL's that are don't want to have. So, I need to do a lot of Google-ing what to do next to just have relevant links.

```python
# Accidently do something different. My intention was to find just relevant links, 
# however this result is nice for me as well. Ok, still need to get those links.
products = soup.find(class_='activeProducts')
products_links = products.find_all('a')
for link in products_links:
    print(link.text)
```

```python
import requests, typing
class Camera(typing.NamedTuple):
  name:str
  quicklook:str
  price:str

from bs4 import BeautifulSoup as soup
d = soup(requests.get('https://www.dpreview.com/products/cameras/all?page=1').text, 'html.parser')
headers = [['div', 'name'], ['div', 'specs'], ['div', 'prices']]
vals = [[getattr(i.find(a, {'class':b}), 'text', 'N/A') for a, b in headers] for i in d.find_all('td', {'class':'product'})]
final_result = [Camera(*i) for i in vals]
```

```python
print(final_result)
```

```python
print(final_result[0].name)
```
